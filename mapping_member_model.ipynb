create a schema for the table according to our requirement

```

%sql
--schema creation
CREATE OR REPLACE TABLE TARGET_TABLE_INDIVIDUAL(
  source_id INT,
  subscriber_id INT,
  first_name VARCHAR(255),
  middle_name VARCHAR(255),
  last_name VARCHAR(255),
  prefix_name VARCHAR(255),
  suffix_name VARCHAR(255),
  name VARCHAR(255),
  record_source VARCHAR(255),
  record_created_ts TIMESTAMP,
  is_verified BOOLEAN,
  addresses ARRAY<STRUCT<address_type:string,address_line_1:string,address_line_2:STRING,city:string,state_provience:string,postal_code:string,zip_code_extension:string,country:string>>,
  phones ARRAY<STRUCT<phone_type:string,number:string>>,
  email VARCHAR(255),
  privacy_preference BOOLEAN,
  national_id VARCHAR(255),
  gender VARCHAR(255),
  marital_status VARCHAR(255),
  date_of_birth DATE,
  year_of_birth VARCHAR(255),
  deceased_ind BOOLEAN,
  deceased_age INT,
  deceased_date DATE,
  languages ARRAY<STRING>,
  employment struct<employer_name:string,employer_role:string,employee_status:string,employee_hiredate:date>,
  additional_source_value map<string,string>
);
```

insert the records into the target table 'target_table_individual' using CTE to adjust multivalue attribute under the same id

```
%sql

--using CTE
with address_cte as 
  (select a.id, array_agg(struct(a.address_type, a.address_line_1, a.address_line_2, a.city, a.state,
    (case 
      when len(a.zipcode) = 10 then split(a.zipcode, '-')[1]
      when len(a.zipcode) = 4 then a.zipcode
      else null 
    end) /*postal code*/,
    (case 
      when len(a.zipcode) = 10 then split(a.zipcode, '-')[0]
      when len(a.zipcode) = 5 then a.zipcode
      else null 
    end) /*zip code*/,
    'US')) as addresses
  from address_as a group by a.id),

----phone
  phone_cte as 
  (select c.id, 
    array_agg(struct(c.usage_type, c.phone)) as contacts 
  from contact_as c group by c.id),

-----date
  date_cte as 
  (select  
    d.id, 
    to_date((replace(d.date_of_birth, '/', '-')), 'M-d-yyyy') as  date_of_birth, 
    to_date((replace(d.deceased_date, '/', '-')), 'M-d-yyyy') as deceased_date,
    to_date((replace(d.job_hiredate, '/', '-')), 'M-d-yyyy') as job_hiredate
  from details_as d)


--using multiple source tables 
insert into target_table_individual
select 
  h.id as source_id, 
  h.insurer_id as subscriber_id,
  d.first_name as first_name, 
  d.middle_name as middle_name, 
  d.last_name as last_name,
  (case 
    when d.gender = 'M' then 'Mr.'
    when d.gender = 'F' and d.marital_status in ('Single', 'Divorced', null) then 'Ms.'
    when d.gender = 'F' and d.marital_status in ('Married', 'Widowed') then 'Mrs.'
    else 'Unknown' 
  end) as prefix_name,
  null as suffix_name,
  (case
    when d.middle_name is null 
    then concat(d.first_name,' ',d.last_name)
    else concat(d.first_name,' ',d.middle_name,' ',d.last_name) 
  end) as name,
  'Nova' as record_source,
  current_timestamp() as record_timestamp,
  false as is_verified,
  ac.addresses as addresses,
  p.contacts as phones,
  (case
    when d.email regexp '^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,}$' then d.email
    else null
  end) as email, 
  false as privacy_preference,
  (case 
    when len(replace(d.ssn, '-','')) > 9 or len(replace(d.ssn, '-','')) < 9 then null
    else concat(substr(d.ssn, 1, 3),'-', substr(replace(d.ssn, '-',''), 4,2),'-',substr(replace(d.ssn, '-',''), 6))
  end) as national_id, 
  d.gender as gender, 
  d.marital_status as marital_status,
  (case 
    when dc.date_of_birth > current_date() then current_date()
    else dc.date_of_birth
  end) as date_of_birth, 
  (case 
    when dc.date_of_birth > current_date() then null
    else year(dc.date_of_birth)
  end) as year_of_birth,
  (case
    when dc.deceased_date is not null then true
    else false 
  end) as deceased_indicator,
  (case 
    when dc.deceased_date > current_date() then year(current_date()) - year(dc.date_of_birth)
    else year(dc.deceased_date) - year(dc.date_of_birth)
  end) as deceased_age,
  (case 
    when dc.deceased_date > current_date() then current_date()
    else dc.deceased_date
  end) as deceased_date,
  (case 
    when d.spoken_language_1 is null or d.spoken_language_2 is null then array(coalesce(d.spoken_language_1,d.spoken_language_2))
    when d.spoken_language_2 = d.spoken_language_1 then array(d.spoken_language_2)
    else array(d.spoken_language_1, d.spoken_language_2)
  end) as languages,
  struct(d.company, d.job_role, 
  (case 
    when dc.deceased_date is null then 'Active'
    else 'Inactive'
  end) as status, 
  (case 
  when dc.date_of_birth > dc.job_hiredate then null
  when dc.job_hiredate > dc.deceased_date then null
  else dc.job_hiredate
  end)) as employment,
  map('relation with subscriber', h.relationship) as additional_source_value
from
header_as h
inner join details_as d on d.id = h.id
inner join date_cte dc on dc.id = h.id
left join phone_cte p on p.id = h.id
inner join address_cte ac on ac.id = h.id;
```
